## Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Sender GSTIN'] = df['Sender GSTIN'].fillna(0)
df['Sender Signature']=df['Sender Signature'].fillna('No',inplace=True)
df['Value Added Services']=df['Value Added Services'].fillna('Paid',inplace=True)
df['Recipient GSTIN'] = df['Recipient GSTIN'].fillna(0)
df['Company Stamp'] = df['Company Stamp'].fillna('No')
df['Receiver Signature'] = df['Receiver Signature'].fillna('No')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Sender Date'] = pd.to_datetime(df['Sender Date'])
df['Expiry Date'] = pd.to_datetime(df['Expiry Date'])
df['Receive Date'] = pd.to_datetime(df['Receive Date'])
```

```python
df.info()
```

```python
df['delivery_days'] = (df['Receive Date'] - df['Sender Date']).dt.days
```

```python
df['month'] = df['Sender Date'].dt.month
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.rename(columns={
    "Actual Wt":"Actual_Weight",
    "Volumetric Wt":"Volumetric_Weight",
    "Chargeable Wt":"Chargeable_Weight",
    "Total Amount":"Total_Amount",
    "Mode of Payment":"Payment_Mode"
}, inplace=True)

df
```
